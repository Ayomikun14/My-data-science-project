From the brief statisical summary. if you look at the category, you would see that there are 9 unique  within categories the columns, let us see what they are.

```python
Category_freq = food["Category"].value_counts()
print(food["Category"].unique())
Category_freq.plot(kind="barh", color = sns.color_palette("hls", 8))
```

The unique features that make up the Category columns are: **'Breakfast' 'Beef & Pork' 'Chicken & Fish' 'Salads' 'Snacks & Sides','Desserts' 'Beverages' 'Coffee & Tea' 'Smoothies & Shakes'**. We can also see that Coffee and Tea has the highest count amongst the 9 categories

Let's examine the categories with the most calories 

```python
sns.barplot(x=food.groupby('Category')['Calories'].mean().index,y=food.groupby('Category')['Calories'].mean().values)
plt.title("total calories of each menu")

plt.ylabel("calories")
plt.xlabel("menu")
plt.xticks(rotation= 90)
plt.show()
```

```python
sns.swarmplot(x = "Calories", y = "Category", data = food)
```

From the barchart Chicken and fish has the highest calories and swarmplot we can see that **Chicken and Fish** is the category with the highest calories in an item. Let's see which item with the highest calories.

```python
food["Calories"].idxmax()
```

```python
max_calories = food.at[82, "Item"]
print("The food with the highest calories is:", max_calories)
```

Next, apart from just calories, let's see which category has the highest in other areas

```python
section = ['Trans Fat', 'Total Fat', 'Cholesterol','Sodium', 'Sugars', 'Carbohydrates', 'Protein']

for x in section:   
    plot = sns.swarmplot(x="Category", y=x, data=food)
    plt.setp(plot.get_xticklabels(), rotation=90)
    plt.title(x)
    plt.show()
```

From the graphs above we can say that:
1. Chicken and Fish has the item with the highest Protein, Sodium and Total Fat
2. Breakfast has the item with the highest Cholesterol level
3. Beef and Pork has the item with the highest Trans Fat
4. Smoothies and shakes has the item with the highest Sugar level and Carbohydrate.

From the Bar Chart, there are 2 food items with the highest Cholesterol amount: Big Breakfast with Hotcake(Large Biscuit ) and Big Breakfast with Hotcake(Regular Size)

```python
Breakfast=food[food["Category"] == "Breakfast"]
ax = Breakfast.plot(kind = "bar",x = "Item",
                                 y = "Cholesterol", figsize=(20, 10)
                          ,width=0.9,color=sns.color_palette('CMRmap',12))
ax.set_title('The Cholesterol levels in Breakfast foods', fontdict = {'fontsize' : 16}, fontweight = "bold")
ax.set_xlabel("Food Items", fontdict = {'fontsize' : 16})
ax.set_ylabel("Calories", fontdict = {'fontsize' : 16})
ax.set_facecolor("white")
ax.tick_params(axis="x", labelsize=14, colors = "black")
ax.set_yticklabels([]) 
ax.axhline(0, color='black') 
for p in ax.patches:
    ax.annotate(str(p.get_height()),(p.get_x()+0.1,p.get_height()+0.01))
```

Double Quarter Pounder with Cheese for the Beef and Pork Food Category has the highest level of Trans fat as shown in the bar chart below.

```python
Beef_and_Pork =food[food["Category"] == "Beef & Pork"]
ax = Beef_and_Pork.plot(kind = "bar",x = "Item",
                                 y = "Trans Fat", figsize=(20, 10)
                          ,width=0.9,color=sns.color_palette('Paired',12))
ax.set_title('The Trans Fat levels in Beef and Pork foods', fontdict = {'fontsize' : 16}, fontweight = "bold")
ax.set_xlabel("Food Items", fontdict = {'fontsize' : 16})
ax.set_ylabel("Trans Fat", fontdict = {'fontsize' : 16})
ax.set_facecolor("white")
ax.tick_params(axis="x", labelsize=14, colors = "black")
ax.set_yticklabels([]) 
ax.axhline(0, color='black') 
for p in ax.patches:
    ax.annotate(str(p.get_height()),(p.get_x()+0.1,p.get_height()+0.01))
```

Furthermore, let's see a heatmap showing the correlation between the columns

```python
f,ax=plt.subplots(figsize=(18,11))
sns.heatmap(food.corr(),annot=True,linewidths=5,fmt='.1f',ax=ax)
```

The heatmap shows the correlation of the columns with each other in a very colourful and interesting manner. Looking at the correlation of Total Fat, Saturated fat, Carbohydrate, Protein and Calories is close to 1. Let's see that in more detail.
